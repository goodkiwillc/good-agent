```python
from pydantic import BaseModel

async with Agent(
    "You are a test agent - provide made up responses when you don't have data."
) as agent:

    class Weather(BaseModel):
        location: str
        temperature: float
        condition: str

    weather = await agent.call(
        "What's the weather like in New York?", response_model=Weather
    )

    print(weather.output.temperature)

    message = await agent.call("Is that typical for this time of year?")

    print(message)
```

```python
# Debate Example

debate_topic = "Is AI beneficial to society?"

prompt = f"You are to debate the following topic: {debate_topic}. You are to take the {{{{position}}}} position. Present your arguments clearly and persuasively."


affirmative = Agent(
    prompt,
    name="affirmative",
    model="gpt-4o-mini",
    context={"position": "affirmative"},
    print_messages=True,
)

negative = Agent(
    prompt,
    name="negative",
    model="gpt-4o-mini",
    context={"position": "negative"},
    print_messages=True,
)

async with (affirmative | negative) as conversation:
    async for message in conversation.execute(max_iterations=4):
        match message:
            case Message(agent=affirmative):
                affirmative.print(message)
            case Message(agent=negative):
                negative.print(message)
```
