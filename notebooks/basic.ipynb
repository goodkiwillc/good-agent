```python
from pydantic import BaseModel

async with Agent("You are a test agent - provide made up responses when you don't have data.") as agent:
    class Weather(BaseModel):
        location: str
        temperature: float
        condition: str

    weather = await agent.call("What's the weather like in New York?", response_model=Weather)

    print(weather)

    message = await agent.call("Is that typical for this time of year?")

    print(message)
```
